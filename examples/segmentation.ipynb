### SAM2

```python
from ezonnx import SAM2, visualize_images
sam2 = SAM2("tiny")
res = sam2.set_image("images/cat.jpg")
# add point with positive or not and its label
res = sam2.set_point([240, 180], True, 0)
res = sam2.set_point([50,180], True, 1)
res = sam2.set_point([400,180], True, 1)
res = sam2.set_point([150, 330], True, 2) 
visualize_images("Masked Image",res.visualized_img)
```

### YOLO-seg

```python
from ezonnx import YOLOSeg
yoloseg = YOLOSeg(onnx_path="/path/to/yolo11n-seg.onnx")
res = yoloseg("images/cat.jpg")
visualize_images("Segmented Image",res.visualized_img)
```
