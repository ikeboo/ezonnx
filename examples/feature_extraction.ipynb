```python
from ezonnx import DinoV2,DinoV3,visualize_imgs

img_path = "images/cat.jpg"

dinov3 = DinoV3("vits16", size=768)
dinov2 = DinoV2("small", size=768)

out_v3 = dinov3(img_path)
out_v2 = dinov2(img_path)

visualize_imgs(["Original image","DinoV3 PCA","DinoV2 PCA"], 
                [out_v3.original_img[...,::-1],out_v3.visualized_img, out_v2.visualized_img])
```

```python
from ezonnx import Siglip2
img_path = "images/cat.jpg"
siglip2 = Siglip2()
res = siglip2(img_path,["two cats","a cat"])
```

```python
from ezonnx import  visualize_images
visualize_images(["Original image","Siglip2 PCA RGB","Siglip2 PCA Grayscale"], 
                [res.original_img[...,::-1],res.pca_img_rgb, res.pca_img_gray])
```
