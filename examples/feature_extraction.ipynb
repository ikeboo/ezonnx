```python
from ezonnx import DinoV2,DinoV3,visualize_images

img_path = "images/cat.jpg"

dinov3 = DinoV3("vits16", size=768)
dinov2 = DinoV2("small", size=768)

out_v3 = dinov3(img_path)
out_v2 = dinov2(img_path)

visualize_images(["Original image","DinoV3 PCA","DinoV2 PCA"], 
                [out_v3.original_img,out_v3.visualized_img, out_v2.visualized_img])
```

```python
from ezonnx import Siglip2, visualize_images
img_path = "images/cat.jpg"
siglip2 = Siglip2()
res = siglip2(img_path,["two cats","a cat","a dog","a car"])
print("Prediction:",res.pred_text)
visualize_images(["Original image","SigLIP2 PCA"], 
                [res.original_img,res.visualized_img])
```
