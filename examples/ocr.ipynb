### PPOCRv5

```python
from ezonnx import PPOCR,visualize_images
ocr = PPOCR("v5-mobile") # or "v5-server" for better accuracy
ret = ocr("images/plate.jpg")
visualize_images(["Target","OCR result"],[ret.original_img,ret.visualized_img])
```

```python
for box in ret.boxed_results:
    print(f"{box.text}")
```
